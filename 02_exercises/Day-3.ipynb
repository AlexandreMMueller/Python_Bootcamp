#### 1. Write a conditional that prints different messages if a bank account balance is below 3,000, between 3,000 and 10,000, or over 10,000.

```python
def account_balance(amount):
    if amount <= 3000:
        print('Your account balance is low.')
    elif amount <= 10000:
        print('Your account balance is sufficient.')
    else:
        print('Your account balance is high.')


account_balance(7500)
#it works!
```

#### 3. Given the list `people`, sort it by people's first name, last name and age. Store the sorted lists as by_first_name, by_last_name, and by_age, respectively.

people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

```python
people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

by_first_name = sorted(people, key = lambda person: person[0])
print(by_first_name)

by_last_name = sorted(people, key = lambda person: person[1])
print(by_last_name)

by_age = sorted(people, key = lambda person: person[2])
print(by_age)

#alternatives by defining functions:
def get_firstname(person):
    return person[0]

def get_lastname(person):
    return person[1]

def get_age(person):
    return person[2]

alt_by_first_name = sorted(people, key = get_firstname)
print(alt_by_first_name)
alt_by_last_name = sorted(people, key = get_lastname)
print(alt_by_last_name)
alt_by_age = sorted(people, key = get_age)
print(alt_by_age)

#Everything works! This is quite interesting. I wanted to experiment using the lambda function. Now I understand how it operates.
```

#### 4. Write a function called `dict_intersect` that takes two dictionaries, d1 and d2, as arguments and returns a set that contains only the keys found in both of the original dictionaries.

```python
d1 = {'Canada': 'Ottawa', 'United States': 'Washington, D.C.', 'England': 'London', 'Germany': 'Berlin'}
d2 = {'Canada': 'Ottawa', 'United States': 'Washington, D.C.', 'Egypt': 'Cairo', 'France': 'Paris'}

def dict_intersect(d1, d2):
    return {k for k in d1 if k in d2}

dict_intersect(d1, d2)
```
