#### 2. Write a program to check if two strings are balanced. For example, strings s1 and s2 are balanced if all the characters in the s1 are present in s2. The character’s position doesn’t matter.

- Ex. s1 = "Yn" , s2 = "PYnative" True, since Yn is in s2
- Ex. s1 = "Ynf", s2 = "PYnative" False, since f is not in s2

*Hint! Iterate each character from a string s1 and check if the current character is present in the string s2.*

```python
s1 = 'nY'
s2 = 'PYnative'

def balancing(s1, s2):
    for characters in s1:
        if not characters in s2:
            return False
    return True 
balancing(s1,s2)

#do note, that claude helped me here in having me recognize that I could simply say
# 'characters' for human knowledge. Since, in python, we ned to use a variable name. 
# I just used 'characters' as the variable name for the procedure, and it worked!
#Is there another way in doing this?
```
