#### 1. Following the function design recipe, define a function that converts kilometers into miles. (Assume there are 1.6 kilometers in a mile.)

```python
def km_to_m(km):
    '''
    Converting Kilometres to Miles.

    Parameters:
    km (float): the distance in kilometres.

    Returns:
    m (float): the distance in miles.
    '''
    return (km * 0.6)

#this is a test#
km_to_m(10)
#it works!
```
