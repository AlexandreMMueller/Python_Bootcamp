```python
def how_cold():
  temp = int(input("What is the temperature in degrees C?"))

  if temp < 0:
    print("It is freezing cold!")
  elif temp <= 15:
    print("It is cold!")
  elif temp <= 30:
    print("It is hot!")
  elif temp <= 45:
    print("It is really hot!")
  else:
    print("It is extremely hot!")
```

```python
how_cold()
```
