```python
def area_circle(radius):
  """
  Caculates the area of a circle given its radius in cm.
  Parameters:
  radius (float): Radius in cm
  Return:
  float: area of a circle with its radius in cm*2
  """
  return (radius **2) * 3.1415
```

```python
area_circle(10)
```

```python
def target_gpa(credits, cgpa, potential_credits, target = 3.5):
  """
  Takes the current CGPA and number of credits as well as potential number of future credits
  and the desired target GPA and returns the average grade needed in all potential credits to acheive target
  """
  return ((target * (credits + potential_credits) - credits * cgpa) / potential_credits)
```

```python
target_gpa(1,3,1)
```

```python
target_gpa(1,3.5,1)
```
